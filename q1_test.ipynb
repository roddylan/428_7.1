```python
import numpy as np
import cv2
# from mat4py import loadmat
import os, shutil
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
mat = sio.loadmat("HouseTallBasler64.mat")
seq = mat["mexVims"]
W = mat["W"]*2
N = mat["NrFrames"].item()
```

```python
test = seq[:,:,0]
test_x = W[0,:]
test_y = W[N,:]
# seq.shape
plt.imshow(test, cmap='gray')
plt.scatter(test_y, test_x)
```

```python
try:
    os.mkdir("test")
except:
    shutil.rmtree("test")
    os.rmdir("test")
    os.mkdir("test")

for i in range(N):
    # test = seq[:,:,i][::,::-1]
    test = seq[:,:,i]
    test_x = W[i,:]
    test_y = W[i + N,:]
    # seq.shape
    plt.imshow(test, cmap='gray')
    plt.scatter(test_y, test_x)
    plt.savefig(f"test/{i:002}.png")
    plt.close()
```
